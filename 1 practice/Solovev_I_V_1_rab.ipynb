—Å–ª–µ–¥—É—é—â–∏—Ö —Ñ—É–Ω–∫—Ü–∏–π: ùëíùë•+ùë¶, sin(ùë• + ùë¶) , cos(ùë• + ùë¶) , ùë•ùë¶.

```python
import math as m
def calc():
    x = int(input("–≤–≤–µ–¥–∏—Ç–µ —Ö"))
    y = int(input("–≤–≤–µ–¥–∏—Ç–µ y"))
    stroke = input("vvedite +, -, *, /, e, sin, cos, ^")
    if stroke == "+":
        print(f'x+y={x+y}')
    if stroke == "-":
        print(f'x-y={x-y}')
    if stroke == "*":
        print(f'x*y={x*y}')
    if stroke == "/":
        print(f'x/y={x/y}')
    if stroke == "e":
        print(f'e^x+y={pow(m.e(1),x+y)}')
    if stroke == "sin":
        print(f'sin(x+y)={m.sin(x+y)}')
    if stroke == "cos":
        print(f'cos(x+y)={m.cos(x+y)}')
    if stroke == "^":
        print(f'x^y={x**y}')

print(calc())
```
